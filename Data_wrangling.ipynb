```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset= pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data', header = None)
```

```python
dataset.head()
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]

dataset.columns= headers
```

```python
dataset.head()
```

```python
# identify and handle the missing values


dataset.replace('?', np.nan, inplace= True)
dataset.head()
```

```python
# evaluating the missing data

```

```python
dataset.isnull().sum()
```

```python
# dealing with the missing data

meanvalue = dataset['normalized-losses'].astype('float').mean(axis= 0)
meanvalue
dataset['normalized-losses'].replace(np.nan, meanvalue, inplace = True)
```

```python
dataset.head()
```

```python
# calculate mean value for bore column

meanbore = dataset['bore'].astype('float').mean(axis= 0)
meanbore
```

```python
dataset['bore'].replace(np.nan, meanbore, inplace= True)
dataset.head()
```

```python
# to see values present in particular column

dataset['stroke'].value_counts()
```

```python
dataset['num-of-doors'].value_counts()
```

```python
# to calculate most common type

dataset['num-of-doors'].value_counts().idxmax()
```

```python
#replace the missing 'num-of-doors' values by the most frequent
dataset["num-of-doors"].replace(np.nan, "four", inplace=True)
```

```python
# drop

dataset.dropna(subset = ['price'], axis = 0, inplace = True)

dataset.reset_index(drop = True, inplace = True)
```

```python
# correct data format

dataset.dtypes
```

```python
dataset[['bore','stroke']]= dataset[['bore','stroke']].astype('float')
dataset[["normalized-losses"]] = dataset[["normalized-losses"]].astype("int")
dataset[["price"]] = dataset[["price"]].astype("float")
dataset[["peak-rpm"]] = dataset[["peak-rpm"]].astype("float")
```

```python
dataset.dtypes
```

```python
# data standardization: Standardization is the process of transforming data into a common format, allowing the researcher to make the meaningful comparison.

dataset.head()
```

```python
dataset['city-mpg']= 235/dataset['city-mpg']

dataset.head()
```

```python
dataset['highway-mpg'] = 235/dataset['highway-mpg']
```

```python
dataset.rename(columns = {"highway-mpg" :"highway-L/100km"}, inplace = True)
```

```python
dataset.head()
```

data normalization
Normalization is the process of transforming values of several variables into a similar range. Typical normalizations include

scaling the variable so the variable average is 0
scaling the variable so the variance is 1
scaling the variable so the variable values range from 0 to 1

```python
dataset['length']= dataset['length']/dataset['length'].max()
dataset['width']= dataset['width']/dataset['width'].max()
dataset['height']= dataset['height']/dataset['height'].max()
```

Binning is a process of transforming continuous numerical variables into discrete categorical 'bins' for grouped analysis.

```python
# Calculate the mean of 'horsepower' column after converting to float
mean_horsepower = dataset['horsepower'].astype('float').mean(axis=0)

# Replace NaN values in 'horsepower' column with the mean
dataset['horsepower'].replace(np.nan, mean_horsepower, inplace=True)

# Convert the 'horsepower' column to integer type
dataset["horsepower"]=dataset["horsepower"].astype(int, copy=True)

# Display the first few rows to verify the changes
display(dataset.head())
```

```python
dataset["horsepower"]=dataset["horsepower"].astype(int, copy=True)
```

```python
import matplotlib as plt
from matplotlib import pyplot

plt.pyplot.hist(dataset["horsepower"])
plt.pyplot.xlabel('horsepower')
plt.pyplot.ylabel('count')
plt.pyplot.title('horsepower bins')
```

```python
bins = np.linspace(min(dataset['horsepower']), max(dataset['horsepower']),4)
bins
```

```python
group_names = ['low', 'medium', 'high']
```

```python
dataset['horsepower_bin']= pd.cut(dataset['horsepower'],bins,labels = group_names, include_lowest = True)
dataset[['horsepower','horsepower_bin']].head(20)
```

```python
dataset['horsepower_bin'].value_counts()
```

```python
pyplot.bar(group_names, dataset['horsepower_bin'].value_counts())
```

```python
# bins visualisation

plt.pyplot.hist(dataset['horsepower'], bins = 3)
```

```python
# indicator values : An indicator variable (or dummy variable)
#is a numerical variable used to label categories. They are called 'dummies'
# You use indicator variables so you can use categorical variables
# for regression analysis in the later modules.
```

```python
dataset.columns
```

```python
dummy_variable = pd.get_dummies(dataset['fuel-type'])
dummy_variable.head()
```

```python
dummy_variable.rename(columns = {'gas': 'fuel-type-gas','diesel': 'fuel-type-diesel'}, inplace = True)
dummy_variable.head()
```

```python
# merge df and dummy_variable

dataset = pd.concat([dataset, dummy_variable], axis = 1)

dataset.drop('fuel-type',axis = 1, inplace = True)
```
