# Dropping irrelevant columns

```python
df = df.drop(['Engine Fuel Type', 'Market Category', 'Vehicle Style', 'Popularity', 'Number of Doors', 'Vehicle Size'], axis= 1)
df.head(5)
```

# Renaming the columns

```python
df = df.rename(columns = {"Engine HP": "HP", "Engine Cylinders": "Cylinders", "Transmission Type": "Transmission", "Driven_Wheels": "Drive Mode",})
df.head(5)
```

# Dropping the duplicate rows

```python
df.shape
```

```python
df.count()
```

```python
df = df.drop_duplicates()
df.head()
```

```python
df.count()
```

# Dropping the missing or null value

```python
print(df.isnull().sum())
```

```python
df = df.dropna()
df.count()
```

# Detecting Outliers

```python
sns.boxplot(x=df['MSRP'])
```

```python
sns.boxplot(x=df['HP'])
```

```python
sns.boxplot(x=df['Cylinders'])
```

```python
Q1 = df.select_dtypes(include='number').quantile(0.25)
Q3 = df.select_dtypes(include='number').quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
# Select only numeric columns for outlier detection
df_numeric = df.select_dtypes(include='number')

# Calculate Q1, Q3, and IQR for numeric columns
Q1 = df_numeric.quantile(0.25)
Q3 = df_numeric.quantile(0.75)
IQR = Q3 - Q1

# Define the outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers from the numeric columns
df_cleaned = df_numeric[~((df_numeric < lower_bound) | (df_numeric > upper_bound)).any(axis=1)]

# Filter the original DataFrame to keep only the rows that are not outliers in the numeric columns
df = df[df.index.isin(df_cleaned.index)]

df.shape
```

# Plot different features against one anaother (scatter) , against(frequency)

Histogram refers to the frequency of occurrence of variables in an interval.

```python
df.Make.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Number of cars by make")
plt.ylabel('Number of cars')
plt.xlabel('Make');
```

Heat Maps is a type of plot which is necessary when we need to find the dependent variables.

```python

plt.figure(figsize=(10,5))
c = df.select_dtypes(include='number').corr()
sns.heatmap(c,cmap="BrBG",annot=True)
print(c)
```

scatter plots to find the correlation between two variables

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.scatter(df['HP'], df['MSRP'])
ax.set_xlabel('HP')
ax.set_ylabel('MSRP')
plt.show()
```
