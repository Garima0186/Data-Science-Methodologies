Results:

What Pearson Correlation Means:
Correlation Coefficient (r): Measures the strength and direction of a linear relationship between two variables.

Ranges from -1 to +1:

+1: Perfect positive correlation

-1: Perfect negative correlation

0: No linear correlation

P-value: Tests the statistical significance of the correlation.

Typically, a p-value < 0.05 indicates a significant correlation.

Null hypothesis


For each Pearson correlation, the null hypothesis (H₀) is:

“There is no linear relationship between the feature (e.g., RAM_GB) and Price.”

This means the true correlation coefficient is 0.

We reject H₀ when the p-value < 0.05, which means the correlation is statistically significant.

| Feature                | Correlation (r) | P-value      | Significant? | Null Hypothesis (H₀) | Effect on H₀                                                 |
| ---------------------- | --------------- | ------------ | ------------ | -------------------- | ------------------------------------------------------------ |
| **RAM\_GB**            | **0.55**        | **3.68e-20** | ✅ Yes        | No correlation       | **Reject H₀** – RAM is significantly correlated with price   |
| **CPU\_frequency**     | **0.37**        | **5.50e-09** | ✅ Yes        | No correlation       | **Reject H₀** – CPU speed has a significant effect           |
| **Storage\_GB\_SSD**   | **0.24**        | **0.00015**  | ✅ Yes        | No correlation       | **Reject H₀** – SSD size influences price slightly           |
| **Screen\_Size\_inch** | **-0.11**       | **0.089**    | ❌ No         | No correlation       | **Fail to reject H₀** – No significant effect of screen size |
| **Weight\_pounds**     | **-0.05**       | **0.44**     | ❌ No         | No correlation       | **Fail to reject H₀** – Weight doesn’t affect price          |
| **CPU\_core**          | **0.46**        | **7.91e-14** | ✅ Yes        | No correlation       | **Reject H₀** – CPU cores significantly affect price         |
| **OS**                 | **-0.22**       | **0.00057**  | ✅ Yes        | No correlation       | **Reject H₀** – OS type has a significant (negative) effect  |
| **GPU**                | **0.29**        | **6.17e-06** | ✅ Yes        | No correlation       | **Reject H₀** – GPU is significantly associated with price   |
| **Category**           | **0.29**        | **7.23e-06** | ✅ Yes        | No correlation       | **Reject H₀** – Category affects price significantly         |


✅ For 7 out of 9 features, the null hypothesis is rejected, meaning those features have a significant linear relationship with price.

❌ For Screen Size and Weight, the null hypothesis cannot be rejected, so we assume no significant linear effect on price.
